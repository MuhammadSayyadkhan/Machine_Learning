```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
df = pd.read_csv("salary_experience.csv")

df.head()
```

```python
plt.scatter(df["Experience (Years)"],df["Salary ($)"])
plt.xlabel("Experience (Years)")
plt.ylabel("Salary ($)")
```

```python
# separate X and Y
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
X.head()
```

```python
# import train test split 
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 2)
X_train.shape
```

```python
# Apply Linear Regression

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_test.head()
```

```python
# now check the model 
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
m = lr.coef_
b = lr.intercept_
```

```python
print("value final and best value of m = ",m)
print("final and best value of b = ",b)
```

```python
# manually prove of prediction model
m*X_test.iloc[0]+b
```

```python
# Draw the prediction line over data 
plt.scatter(df["Experience (Years)"],df["Salary ($)"])
plt.plot(X_train,lr.predict(X_train),c='r')
plt.xlabel("Experience (Years)")
plt.ylabel("Salary ($)")
plt.show()
```
